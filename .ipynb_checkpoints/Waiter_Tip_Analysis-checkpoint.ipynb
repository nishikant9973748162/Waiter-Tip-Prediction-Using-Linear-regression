## Waiter tip prediction model

#### Data Transformation : Categorical values to numerical ones

```python
data["payment_method"] = data["payment_method"].apply(lambda x: {"Card": 0, "Cash": 1, "Online": 2}.get(x, -1))
data["day"] = data["day"].apply(lambda x: {"Sunday": 0, "Monday": 1, "Friday": 2, "Saturday": 3}.get(x, -1))
data["timing"] = data["timing"].apply(lambda x: {"Lunch": 0, "Dinner": 1}.get(x, -1))
data["group"] = data["group"].apply(lambda x: {"Single": 0, "Family": 1, "Couple": 2, "Friends": 3}.get(x, -1))
data["gender"] = data["gender"].apply(lambda x: {"Male": 0, "Female": 1}.get(x, -1))
data["occasion"] = data["occasion"].apply(lambda x: {"Date": 0, "Birthday Party": 1, "Anniversary": 2, "Festival": 3}.get(x, -1))
data["drunk"] = data["drunk"].apply(lambda x: {"Drunk": 0, "Normal": 1}.get(x, -1))
data.head()
```

#### Spliting the data into training and test sets

```python
x = np.array(data[["total_bill", "payment_method", "day", "timing","group","gender","occasion","drunk"]])
y = np.array(data["tip"])
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y,test_size=0.2,random_state=42)
```

#### Training of model

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')  # or 'median', 'most_frequent', etc.
xtrain = imputer.fit_transform(xtrain)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

#### Features to input = [[total_bill, "payment_method", "day", "timing", "group", "gender,occasion","drunk"]]


```python
features = np.array([[450,1,2,1,0,0,1,1]])
model.predict(features)
```
